```python
#変数varは正の整数とします
import random
var = random.randint(1,30)

#変数varが、3の倍数と5の倍数の場合は「FizzBuzz」を出力
if var % 5 == 0 and var % 3 ==0:
    print("FizzBuzz")
#変数varが、3の倍数の場合は「Fizz」を出力
if var % 3 == 0:
    print("Fizz")
#変数varが、5の倍数の場合は「Buzz」を出力
elif var % 5 == 0:
    print("Buzz")

#変数varが、上記のどれにも該当しない場合
else :
    print(var)

```
